```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data_ch_3 = pd.read_csv('output_interp_ch_3.txt', sep=" ", names = ['Длина отрезка', 'Ошибка'])
data_ch_4 = pd.read_csv('output_interp_ch_4.txt', sep=" ", names = ['Длина отрезка', 'Ошибка'])
data_ch_5 = pd.read_csv('output_interp_ch_5.txt', sep=" ", names = ['Длина отрезка', 'Ошибка'])

data_uni_3 = pd.read_csv('output_interp_uni_3.txt', sep=" ", names = ['Длина отрезка', 'Ошибка'])
data_uni_4 = pd.read_csv('output_interp_uni_4.txt', sep=" ", names = ['Длина отрезка', 'Ошибка'])
data_uni_5 = pd.read_csv('output_interp_uni_5.txt', sep=" ", names = ['Длина отрезка', 'Ошибка'])
```
