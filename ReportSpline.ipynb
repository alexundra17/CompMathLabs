```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('output_spline.txt', sep=" ", names = ['Узлы', 'Ошибка'])
```

```python
data
```

```python
plt.figure(dpi = 100)
plt.grid()
plt.xlabel('log(N)')
plt.ylabel('log(Error)')
plt.plot(np.log(data['Узлы']).values, np.log(data['Ошибка']).values)
k, b = np.polyfit(np.log(data['Узлы']).values, np.log(data['Ошибка']), 1)
plt.text (3.5, 8, f'k ={k}') 
plt.title('Логарифм ошибки интерполяции \n от логарифма количества узлов N')
```
