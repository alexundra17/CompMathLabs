```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data3 = pd.read_csv('data_3.csv', sep=" ", names = ['Ошибка', 'Шаг'])
data4 = pd.read_csv('data_4.csv', sep=" ", names = ['Ошибка', 'Шаг'])
data5 = pd.read_csv('data_5.csv', sep=" ", names = ['Ошибка', 'Шаг'])
```

```python
data3
```

```python
plt.figure(dpi = 100)
plt.grid()
plt.xlabel('log(h)')
plt.ylabel('log(Error)')
plt.plot(np.log(data3['Шаг']).values, np.log(data3['Ошибка']).values, marker = '*')
k, b = np.polyfit(np.log(data3['Шаг']).loc[11:].values, np.log(data3['Ошибка']).loc[11:].values, 1)
print(k)

plt.title(f'Логарифм ошибки дифферинцирования \n от логарифма шага \n k = {k} \n N = 3')
```

```python
plt.figure(dpi = 100)
plt.grid()
plt.xlabel('log(h)')
plt.ylabel('log(Error)')
plt.plot(np.log(data4['Шаг']).values, np.log(data4['Ошибка']).values, marker = '*')
k, b = np.polyfit(np.log(data4['Шаг']).loc[11:].values, np.log(data4['Ошибка']).loc[11:].values, 1)
print(k)

plt.title(f'Логарифм ошибки дифферинцирования \n от логарифма шага \n k = {k} \n N = 4')
```
