```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('output_integrate.txt', sep=" ", names = ['Шаг', 'Ошибка'])
```

```python
data
```

```python
plt.figure(dpi = 100)
plt.grid()
plt.xlabel('log(h)')
plt.ylabel('log(Error)')
plt.plot(np.log(data['Шаг']).values, np.log(data['Ошибка']).values, marker = '*')
k, b = np.polyfit(np.log(data['Шаг']).loc[1:].values, np.log(data['Ошибка']).loc[1:].values, 1)
print(k)

plt.title(f'Логарифм ошибки интегрирования \n от логарифма шага \n k = {k}')
```
